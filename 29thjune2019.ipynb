```python
# factorial of given number
n=int(input())
def factorial(n):
    s=1
    for i in range(1,n+1):
        s=s*i
    print(s)
factorial(n)
```

```python
def countDivisors(i,j,k):
    count=0
    for n in range(i,j+1):
        if n%k==0:
            count+=1
    print(count)
s=input()
s=s.split()
i=int(s[0])
j=int(s[1])
k=int(s[2])
countDivisors(i,j,k)
```

```python
def isprimenum(n):
    for i in range(2,n):
        if (n%i==0):
            return False
    return True
def primenumber(n):
    for i in range(2,n+1):
        if isprimenum(i):
            print(i,end=" ")
n=int(input())
primenumber(n)
```

```python
fruits=['orange','apple','mango','apple','mango']
fruit_dict={}
for fruit in fruits:
    fruit_dict[fruit]=fruit_dict.get(fruit,0)+1
    
print(fruit_dict)
```

```python
len(fruit_dict)
```

```python
for key in fruit_dict.keys():
    print(key,fruit_dict.get(key),sep=":")
```

```python
print(fruit_dict.values(),end=" ")
```
